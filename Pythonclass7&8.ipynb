Arbitorary Position arguments

 *args

```python
import builtins

def work(*numbers):
  return builtins.sum(numbers)
print(work(1,2,3,4,5,6))
```

```python
import builtins

def hasi(*args):
  return builtins.sum(args)

hasi(1,2,3,4)
```
